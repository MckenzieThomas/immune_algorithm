```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D
```

```python
def shestigorb(x, y):
    return (4 - 2.1 * x**2 + (x**4) / 3) * x**2 + x*y + (-4 - 4 * y**2)
```

```python
def ionha(X, Y):
    return (sum([x**2 for x in X]) / sum([y**2 for y in Y]))
```

```python
def paralel(X):
    return sum([i*x**2 for x, i in zip(X, range(1,len(X)+1))])
```

```python
def goldstein(x, y):
    return (1 + (1+x+y)**2 * (19-14*x+3*x**2 - 14*y + 6*x*y +3*y**2)) * (30+ (2*x - 3*y)**2 * (18 - 32*x + 12*x**2 + 48*y - 36*x*y + 27*y**2))
```

```python
def beale(X, Y):
    return (1.5-X+X*Y)**2+(2.25-X+X*Y**2)**2+(2.625-X+X*Y**3)**2  
    
x = np.arange(-4.5, 4.5, 0.1)
y = np.arange(-4.5, 4.5, 0.1)
X, Y = np.meshgrid(x, y)
L = beale(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, L, rstride=1, cstride=1,cmap='jet', edgecolor = 'none')
```
